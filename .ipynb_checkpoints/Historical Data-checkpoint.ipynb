# Historical Data
This is for looking at the older data that Brandon sent us after the first meeting.

### Setup

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import json
import csv
from collections import defaultdict

#This lets us zoom in and move around graphs in this notebook
%matplotlib widget
```

```python
#Get the streets and buildings that we'll want to see on maps:
buildings = None
with open('buildings_cu.json') as f:
    buildings = json.loads(f.read())

streets = None
with open('streets_near_campus.json') as f: #these will have duplicates. There are multiple entries for roads like Broadway
    streets = json.loads(f.read())
```

```python
print("CU Buildings: {}".format(len(buildings)))
print("Near Campus Streets: {} (there are probably duplicates)".format(len(streets)))
```

```python
#Get station information (latitude and longitude, names)
stations = None
with open('station_information.json') as f: #TODO: Use urlopen to get the current one?
    stations = json.loads(f.read())['data']['stations']

print("There are {} Boulder BCycle Stations".format(len(stations)))

#Do some convenience stuff to stations:

#make it a dictionary with ids as keys
temp = stations
stations = {}
for station in temp:
    stations[station['station_id']] = station

#For convenience:
def GetStationID(stationName):
    for stationID in stations:
        if stations[stationID]['name'] == stationName:
            return stationID
    return None
```

```python
#In order to know how many docks each station has, we need to load one status file.
with open('GBFSTest1_StationStatus_Output/1644027738.json') as f:
    status = json.loads(f.read())
    
    for station in status['data']['stations']:
        stations[station['station_id']]['total_docks'] = station['num_bikes_available'] + station['num_docks_available']
```

```python
lats = [stations[station]['lat'] for station in stations]
lons = [stations[station]['lon'] for station in stations]

#plot
plt.figure()
plt.suptitle('Map of All Stations')

for building in buildings:
    x = [pair[0] for pair in building['geometry']['coordinates'][0]]
    y = [pair[1] for pair in building['geometry']['coordinates'][0]]

    plt.fill(x, y, facecolor='xkcd:bright blue', edgecolor='xkcd:cobalt blue')

for street in streets:
    x = [pair[0] for pair in street['geometry']['coordinates']]
    y = [pair[1] for pair in street['geometry']['coordinates']]

    plt.plot(x, y, color='xkcd:grey')

plt.plot(lons, lats, '.', color='xkcd:cherry red')
    
plt.ylabel('lat')
plt.xlabel('lon')
plt.ticklabel_format(useOffset=False)

plt.show()
```

```python
tripsByStart = {}
unknownStations = set()
with open('B-cycleRideMovement4_13.csv', newline='') as f:
    reader = csv.reader(f)
    lines = [row for row in reader]
    
    #for each row
    #Ignore the first row with the labels, and the last row with the totals.
    for i in range(1, len(lines) - 1):
        start = GetStationID(lines[i][1])
        
        #Make sure we actually found the id of the station we're looking for
        if start is None:
            unknownStations.add(lines[i][1])
            continue
        
        tripsByStart[start] = {}
        
        #for each column
        #Ignore the first column with the rank, and the last column with the totals.
        for k in range(2, len(lines[i]) - 1):
            end = GetStationID(lines[0][k])
            
            #make sure there's a station with the name we're looking for
            if end is None:
                unknownStations.add(lines[0][k])
                continue
            
            entry = int(lines[i][k]) if lines[i][k] != "" else 0
            
            tripsByStart[start][end] = entry

print("Couldn't find station ids for:")
print(unknownStations)
```

### Map of Trips
Each route's alpha corresponds to its relative number of trips. Only the most common one has an alpha of 1. Each arrow only has a head on its right side, which lets you see the difference in strength of the route going to and from any particular station.

This older data seems to have several stations that either no longer exist, or have been renambed, and it doesn't seem to have most of the stations on campus.

TODO: Maybe give each station its own unique color, and color the arrows starting at that station accordingly? Alternatively, since we aren't actually drawing the real world path that people took anyway, the arrows could curve away from the line between the stations a little.

```python
lats = [stations[station]['lat'] for station in stations]
lons = [stations[station]['lon'] for station in stations]

maxTripCount = 0
for start in tripsByStart.values():
    for entry in start.values():
        if entry > maxTripCount:
            maxTripCount = entry

print("The most trips on any particular route was {}.".format(maxTripCount))

#plot
plt.figure()
plt.suptitle('Map of Routes')
plt.title('Shaded by Number of Trips')

for building in buildings:
    x = [pair[0] for pair in building['geometry']['coordinates'][0]]
    y = [pair[1] for pair in building['geometry']['coordinates'][0]]

    plt.fill(x, y, facecolor='xkcd:bright blue', edgecolor='xkcd:cobalt blue')

for street in streets:
    x = [pair[0] for pair in street['geometry']['coordinates']]
    y = [pair[1] for pair in street['geometry']['coordinates']]

    plt.plot(x, y, color='xkcd:grey')

plt.plot(lons, lats, '*', color='xkcd:gold', markeredgecolor='black', markersize=10, markeredgewidth=0.5)

#plot the trips:
for startID in tripsByStart:
    for endID in tripsByStart[startID]:
        num = tripsByStart[startID][endID]
        
        x = stations[startID]['lon']
        y = stations[startID]['lat']
        dx = stations[endID]['lon'] - x
        dy = stations[endID]['lat'] - y
        a = num/maxTripCount
        
        if startID != endID:
            #NOTE: it seems like it actually draws the arrow with the opposite side as specified by shape.
            plt.arrow(x, y, dx, dy, width=0.0001, head_width=0.001, shape='left', color='xkcd:cherry red', alpha=a)
        else:
            #if this is the number that started and ended at the same station, we just draw a dot on that station.
            plt.plot(x, y, '.', color='xkcd:cherry red', markersize=5)
    
plt.ylabel('lat')
plt.xlabel('lon')
plt.ticklabel_format(useOffset=False)

plt.show()
```

```python
#figure out the number of total trips involving (starting or ending at) each station
tripsPerStationTotal = defaultdict(int)

for startID in tripsByStart:
    for endID in tripsByStart[startID]:
        tripsPerStationTotal[startID] += tripsByStart[startID][endID]
        tripsPerStationTotal[endID] += tripsByStart[startID][endID]
```

```python
tripsPerStationPerDock = {}
for stationID in tripsPerStationTotal:
    tripsPerStationPerDock[stationID] = tripsPerStationTotal[stationID]/stations[stationID]['total_docks']
    print("{} trips per dock: {}".format(stationID, tripsPerStationPerDock[stationID]))

print("\nMedian trips per dock: {}".format(np.median(list(tripsPerStationPerDock.values()))))
print("Mean trips per dock: {}".format(np.mean(list(tripsPerStationPerDock.values()))))
```

### Map of Stations

I think this lets us see which stations are getting used more or less than we'd expect based on their size. Those that are used more or less often than we'd expect might be more volatile (meaning they might be more likely to always be either completely full or completely empty), and we might want to look into them.

If a station is overused, we could try to figure out where the people who are using it need to go, and add stations to give them a more direct way of doing so. (for example, say the will vill and c4c stations are overused. If we find out that a lot of those people really want to go from will vill to farrand field, and the c4c just happens to be the nearest station, we could add a station at farrand field, and that would take some pressure of the c4c station.)

```python
maxTripsPerDock = max(tripsPerStationPerDock.values())
minTripsPerDock = min(tripsPerStationPerDock.values())

print("min, max Trips per dock: [{}, {}]".format(minTripsPerDock, maxTripsPerDock))

#plot
plt.figure()
plt.suptitle('Map of Stations, colored by Relative Trips Per Dock')
plt.title('Red=Max, Green=Median, Blue=Min')

for building in buildings:
    x = [pair[0] for pair in building['geometry']['coordinates'][0]]
    y = [pair[1] for pair in building['geometry']['coordinates'][0]]

    plt.fill(x, y, facecolor='xkcd:bright blue', edgecolor='xkcd:cobalt blue')

for street in streets:
    x = [pair[0] for pair in street['geometry']['coordinates']]
    y = [pair[1] for pair in street['geometry']['coordinates']]

    plt.plot(x, y, color='xkcd:grey')


medianTripsPerDock = np.median(list(tripsPerStationPerDock.values()))
for stationID in tripsPerStationTotal:
    r = 0 if tripsPerStationPerDock[stationID] < medianTripsPerDock else (tripsPerStationPerDock[stationID]-medianTripsPerDock)/(maxTripsPerDock-medianTripsPerDock)
    b = 0 if tripsPerStationPerDock[stationID] > medianTripsPerDock else (medianTripsPerDock-tripsPerStationPerDock[stationID])/(medianTripsPerDock-minTripsPerDock)
    g = 1-b if tripsPerStationPerDock[stationID] < medianTripsPerDock else 1-r
    
    c = (r, g, b)
    plt.plot(stations[stationID]['lon'], stations[stationID]['lat'], '*', color=c, markeredgecolor='black', markersize=10, markeredgewidth=1)
    
plt.ylabel('lat')
plt.xlabel('lon')
plt.ticklabel_format(useOffset=False)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy
import scipy
import os
import json
from datetime import datetime
#import ipympl

%matplotlib widget
```

```python
buildings = None
with open('Building_Footprints.geojson') as f: #The boulder building footprints from Boulder's Open Data
    buildings = json.loads(f.read())
```

```python
buildings['features'][0]
```

```python
plt.figure()

plt.suptitle("Folsom Stadium")

for building in buildings['features']:
    if building['properties']['SHORTNAME'] == "Folsom Stadium":
        x = [pair[0] for pair in building['geometry']['coordinates'][0]]
        y = [pair[1] for pair in building['geometry']['coordinates'][0]]
        
        #print(x, y)

        plt.fill(x, y)

plt.xlabel('lat')
plt.xlabel('lon')

plt.show()
```

```python
plt.figure()

plt.suptitle("Boulder Buildings")
plt.title("Zoomed in on Main Campus")

for building in buildings['features']:
    x = [pair[0] for pair in building['geometry']['coordinates'][0]]
    y = [pair[1] for pair in building['geometry']['coordinates'][0]]

    #print(x, y)

    plt.fill(x, y, facecolor='xkcd:bright blue', edgecolor='xkcd:cobalt blue')

plt.ylim(40.00377324157075, 40.008464589267255)
plt.xlim(-105.27197458731145, -105.26013291890786)
    
plt.xlabel('lat')
plt.xlabel('lon')

plt.show()
```

```python
campusBuildings = [building for building in buildings['features'] if building['properties']['LONGNAME'] and ("University of Colorado" in building['properties']['LONGNAME'] or "Williams Village" in building['properties']['LONGNAME'])]
```

```python
plt.figure()

plt.suptitle("Boulder Buildings")
plt.title("Autofiltered Campus Buildings Only")

for building in campusBuildings:
    x = [pair[0] for pair in building['geometry']['coordinates'][0]]
    y = [pair[1] for pair in building['geometry']['coordinates'][0]]

    #print(x, y)

    plt.fill(x, y, facecolor='xkcd:bright blue', edgecolor='xkcd:cobalt blue')

#plt.ylim(40.00377324157075, 40.008464589267255)
#plt.xlim(-105.27197458731145, -105.26013291890786)
    
plt.xlabel('lat')
plt.xlabel('lon')

plt.show()
```

```python
#Get station information (latitude and longitude, names)
stations = []
with open('station_information.json') as f: #TODO: Use urlopen to get the current one?
    stations = json.loads(f.read())['data']['stations']

print("{} Stations".format(len(stations)))
```

```python
#Do some convenience stuff to stations:

#make it a dictionary with ids as keys
temp = stations
stations = {}
for station in temp:
    stations[station['station_id']] = station
```

```python
lats = [stations[station]['lat'] for station in stations]
lons = [stations[station]['lon'] for station in stations]

#plot
plt.figure()
plt.suptitle('Bad (not terrible) map of All Stations')

plt.plot(lons, lats, '.', color='xkcd:cherry red') #assume the Earth is flat.

for building in campusBuildings:
    x = [pair[0] for pair in building['geometry']['coordinates'][0]]
    y = [pair[1] for pair in building['geometry']['coordinates'][0]]

    #print(x, y)

    plt.fill(x, y, facecolor='xkcd:bright blue', edgecolor='xkcd:cobalt blue')
    
plt.xlabel('lat')
plt.xlabel('lon')

plt.show()
```

```python
streets = None
with open('Street_Centerlines.geojson') as f: #The street centerlines from Boulder's Open Data
    streets = json.loads(f.read())
```

```python
streets['features']
```

```python
streets['features'][0]['geometry']['coordinates'][0]
```

```python
plt.figure()

plt.suptitle("Boulder Streets")
plt.title("Zoomed in on Main Campus")

for street in streets['features']:
    x = [pair[0] for pair in street['geometry']['coordinates']]
    y = [pair[1] for pair in street['geometry']['coordinates']]

    #print(x, y)

    plt.plot(x, y, color='xkcd:grey')

plt.ylim(40.00377324157075, 40.008464589267255)
plt.xlim(-105.27197458731145, -105.26013291890786)
    
plt.xlabel('lat')
plt.xlabel('lon')

plt.show()
```

```python
plt.figure()

plt.suptitle("Boulder Buildings and Streets")
plt.title("Zoomed in on Main Campus")

for building in buildings['features']:
    x = [pair[0] for pair in building['geometry']['coordinates'][0]]
    y = [pair[1] for pair in building['geometry']['coordinates'][0]]

    #print(x, y)

    plt.fill(x, y, facecolor='xkcd:bright blue', edgecolor='xkcd:cobalt blue')

for street in streets['features']:
    x = [pair[0] for pair in street['geometry']['coordinates']]
    y = [pair[1] for pair in street['geometry']['coordinates']]

    #print(x, y)

    plt.plot(x, y, color='xkcd:grey')

plt.ylim(40.00377324157075, 40.008464589267255)
plt.xlim(-105.27197458731145, -105.26013291890786)
    
plt.xlabel('lat')
plt.xlabel('lon')

plt.show()
```

```python
#Better Building and Street filter, near campus. Includes Main Campus, East Campus, and Will Vill

latMin = 39.99750
latMax = 40.01475
lonMin = -105.2750
lonMax = -105.2400

nearCampusBuildings = []
for building in buildings['features']:
    for point in building['geometry']['coordinates'][0]:
        try:
            if point[0] >= lonMin and point[0] <= lonMax and point[1] >= latMin and point[1] <= latMax:
                nearCampusBuildings.append(building)
                break
        except:
            pass

#print(nearCampusBuildings)
        
nearCampusStreets = []
for street in streets['features']:
    for point in street['geometry']['coordinates']:
        try:
            if point[0] >= lonMin and point[0] <= lonMax and point[1] >= latMin and point[1] <= latMax:
                nearCampusStreets.append(street)
                break
        except:
            pass
```

```python
print("Near Campus Buildings: {}".format(len(nearCampusBuildings)))
print("Near Campus Streets: {}".format(len(nearCampusStreets)))
```

```python
plt.figure()

plt.suptitle("Near Campus Buildings and Streets")
#plt.title("Zoomed in on Main Campus")

for building in nearCampusBuildings:
    x = [pair[0] for pair in building['geometry']['coordinates'][0]]
    y = [pair[1] for pair in building['geometry']['coordinates'][0]]

    plt.fill(x, y, facecolor='xkcd:bright blue', edgecolor='xkcd:cobalt blue')

for street in nearCampusStreets:
    x = [pair[0] for pair in street['geometry']['coordinates']]
    y = [pair[1] for pair in street['geometry']['coordinates']]

    plt.plot(x, y, color='xkcd:grey')
    
plt.xlabel('lat')
plt.xlabel('lon')

plt.show()
```

```python
cuBuildings = []
for building in nearCampusBuildings:
    if building['properties']['LONGNAME'] is not None:
        cuBuildings.append(building)
```
